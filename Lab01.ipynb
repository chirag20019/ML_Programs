```python
import pandas as pd

# Load the dataset
file_path = 'Adaniports.csv'
data = pd.read_csv(file_path)

# Display the first few rows
print(data.head())
```

```python
# Check data structure
print(data.info())
```

```python
# Statistical summary
print(data.describe())
```

```python
# Fill missing values (example)
data.fillna(method='ffill', inplace=True)
```

```python
import matplotlib.pyplot as plt
# Example: Plot Closing Prices over Time
data['Date'] = pd.to_datetime(data['Date'])  # Ensure 'Date' is a datetime object
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Close'], label='Close Price')
plt.title('Adani Ports Stock Closing Prices (2000-2021)')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

```python
import seaborn as sns
# Distribution plot for a numeric column (e.g., 'Close')
sns.histplot(data['Close'], kde=True)
plt.title('Distribution of Closing Prices')
plt.show()
```

```python
# Box plot for 'Close'
sns.boxplot(x=data['Close'])
plt.title('Boxplot for Closing Prices')
plt.show()
```
